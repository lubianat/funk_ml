```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
import re
from tqdm import tqdm


def return_lyrics_dataframe(artist_name, artist_style):
    
    artist_query = requests.get("https://www.vagalume.com.br/" + 
                                  artist_name +
                                  "/index.js").json()
    
    API_URL = 'https://api.vagalume.com.br'
    API_LYRIC_SEARCH_URL = API_URL + '/search.php'
    
    artist_lyrics = json_normalize(artist_query['artist']['toplyrics']['item'])['desc']

    artist_dataframe = pd.DataFrame(columns = ['name', 'text', 'artist', 'style'])

    for music in tqdm(artist_lyrics):
        params = {
            'art': artist_name,
            'mus': music
        }
        try:
            lyrics_query = requests.get(API_LYRIC_SEARCH_URL, params = params)
        except:
            try:
                lyrics_query = requests.get(API_LYRIC_SEARCH_URL, params = params)
            except:
                try:
                    lyrics_query = requests.get(API_LYRIC_SEARCH_URL, params = params)
                except:
                    continue
        df = json_normalize(lyrics_query.json()['mus'])
        df['artist'] = artist_name
        df['style'] = artist_style
        df = df[['name', 'text', 'artist', 'style']]
        artist_dataframe = artist_dataframe.append(df)
    return(artist_dataframe)
```

```python

funk_carioca = ['mc-th','gaiola-das-popozudas', 'kevin-o-chris','ludmilla', 'nego-do-borel', 'dennis-dj', 'mc-ysa', 'mc-carol', 'mr-catra','valesca-popozuda']

funk_paulista = ['mc-don-juan','mc-tati-zaqui','mc-kevinho','mc-livinho','mc-wm','mc-kekel', 'mc-fioti', 'mc-bin-laden', "mc-zaac"]

funk_mineiro = ['mc-l-da-vinte']
```

```python
carioca_dataframe = pd.DataFrame(columns = ['name', 'text', 'artist', 'style'])


import time

for artista in funk_carioca:
    print(artista)
    style = 'funk_carioca'
    name = artista 
    try:
        dataframe_for_this_artist = return_lyrics_dataframe(name, style)
    except:
        print('marcando 30 segundos')
        time.sleep(30)
        try:
            dataframe_for_this_artist = return_lyrics_dataframe(name, style)
        except:
            print('failed mesmo')
    carioca_dataframe = carioca_dataframe.append(dataframe_for_this_artist)
    print('marcando 30 segundos')
    time.sleep(30)
    
carioca_dataframe.to_csv('funks_cariocas.csv')
```
